```python
import numpy as np
import matplotlib.pyplot as plt

r = 1
theta = np.linspace(0, 2*np.pi, 1000)

# circulito
x_circle = r * np.cos(theta)
y_circle = r * np.sin(theta)

n_points = 1000
t = np.linspace(0, 4*np.pi, n_points)
x_involute = r * (np.cos(t) + t * np.sin(t))
y_involute = r * (np.sin(t) - t * np.cos(t))

plt.figure(figsize=(8, 8))
plt.plot(x_circle, y_circle, label='Círculo')
plt.plot(x_involute, y_involute, label='Involuta')

plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Círculo e Involuta')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#elipse
a = 3  # Semieje mayor
b = 2  # Semieje menor
theta = np.linspace(0, 2*np.pi, 1000)

x_ellipse = a * np.cos(theta)
y_ellipse = b * np.sin(theta)

t = np.linspace(0, 2*np.pi, 1000)
x_involute = a * (np.cos(t) + t * (-np.sin(t)))
y_involute = b * (np.sin(t) + t * np.cos(t))

plt.figure(figsize=(8, 8))
plt.plot(x_ellipse, y_ellipse, label='Elipse')
plt.plot(x_involute, y_involute, label='Involuta')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Elipse e Involuta')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# astroida
a = 1
theta = np.linspace(0, 2*np.pi, 1000)

x_astroid = a * np.cos(theta)**3
y_astroid = a * np.sin(theta)**3

t = np.linspace(0, 2*np.pi, 1000)
x_involute = (a**2) * (3 * np.cos(t) - np.cos(3 * t)) / 4
y_involute = (a**2) * (3 * np.sin(t) - np.sin(3 * t)) / 4

plt.figure(figsize=(8, 8))
plt.plot(x_astroid, y_astroid, label='Astroida')
plt.plot(x_involute, y_involute, label='Involuta')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Astroida e Involuta')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# cardioide
a = 1  # Radio
theta = np.linspace(0, 2*np.pi, 1000)


x_cardioid = a * (2 * np.cos(theta) - np.cos(2*theta))
y_cardioid = a * (2 * np.sin(theta) - np.sin(2*theta))

t = np.linspace(0, 2*np.pi, 1000)
x_involute = a * (2*np.sin(t) - np.sin(2*t) - 2*t*np.cos(t))
y_involute = a * (-2*np.cos(t) + np.cos(2*t) + 2*t*np.sin(t))

plt.figure(figsize=(8, 8))
plt.plot(x_cardioid, y_cardioid, label='Cardioide')
plt.plot(x_involute, y_involute, label='Involuta')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Cardioide e Involuta')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# catenaria
a = 1  # Constante de la catenaria
x = np.linspace(-2, 2, 1000)

# Coordenadas de la catenaria
y_catenary = a * np.cosh(x / a)

arc_length = a * np.sinh(x/a)
x_involute = x - y_catenary * (arc_length - arc_length[0]) / a
y_involute = y_catenary - (arc_length - arc_length[0])

plt.figure(figsize=(8, 8))
plt.plot(x, y_catenary, label='Catenaria')
plt.plot(x_involute, y_involute, label='Involuta')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Catenaria e Involuta')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
